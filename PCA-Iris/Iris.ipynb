```python
x = df.iloc[:,:4]
y = df.iloc[:,-1:].values
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(x)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
pca_X = pca.fit_transform(X)
pca_X
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score,confusion_matrix
def heatmapGraph(cm,title,acc_score):
    plt.style.use("ggplot")
    sns.heatmap(cm,annot=True,fmt=".2f")
    plt.title("{} Accuracy Score:{}".format(title,acc_score))
    plt.show()  
def algorithmKNN(x,y):
    scores = []
    for i in range(1,12):
        knn = KNeighborsClassifier(n_neighbors=i)
        knn.fit(x,y)
        y_pred = knn.predict(x)
        scores.append(accuracy_score(y, y_pred))
    heatmapGraph(confusion_matrix(y,y_pred),"KNN",accuracy_score(y,y_pred))
    return scores
print(algorithmKNN(X,y))
```

```python
print(algorithmKNN(pca_X,y))
```

PCA finds the best algorithm that separates the data by minimizing the number of columns and thus finds the prediction result with a small amount of error compared to the normal prediction results.

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X,y)
y_pred = log_reg.predict(X)
heatmapGraph(confusion_matrix(y,y_pred),"Log.Reg",accuracy_score(y,y_pred))
```

```python
log_reg = LogisticRegression()
log_reg.fit(pca_X,y)
y_pred = log_reg.predict(pca_X)
heatmapGraph(confusion_matrix(y,y_pred),"Log.Reg PCA",accuracy_score(y,y_pred))
```

```python
from sklearn.svm import SVC
svm = SVC(kernel="rbf")
svm.fit(X,y)
y_pred = svm.predict(X)
# PCA 
svm_pca = SVC(kernel="rbf")
svm_pca.fit(pca_X,y)
y_pred_pca = svm_pca.predict(pca_X)
plt.subplot(1,2,1)
heatmapGraph(confusion_matrix(y,y_pred),"Log.Reg",accuracy_score(y,y_pred))
plt.subplot(1,2,2)
heatmapGraph(confusion_matrix(y,y_pred_pca),"Log.Reg PCA",accuracy_score(y,y_pred_pca))
```
